```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')

import missingno
```

```python
base_dados = pd.read_csv(r"C:\Users\Joao Paulo\OneDrive\Documents\Python_para_Analise_de_Dados_Data_Viking\Salary_Data.csv")
base_dados.head()
```

```python
base_dados = base_dados.rename(columns={
    'YearsExperience': 'years_of_experience',
    'Salary': 'salary'
})
```

```python
# Análise da distribuição da renda

plt.figure(figsize=(10,6))
plt.title('Distribuição da renda')
sns.kdeplot(data=base_dados['salary'], color='#D41652');
```

```python
# Análise da distribuição da experiência

plt.figure(figsize=(10,6))
plt.title('Distribuição da experiência')
sns.kdeplot(data=base_dados['years_of_experience'], color='#831DE8');
```

```python
# Distribuição
plt.figure(figsize=(10,6))
sns.boxplot(data=base_dados, x='salary');
```

```python
# Distribuição
plt.figure(figsize=(10,6))
sns.boxplot(data=base_dados, x='years_of_experience');
```

```python
# Distribuição
plt.figure(figsize=(10,6))
plt.title('Relação entre tempo de experiência e salário', fontsize=18)
sns.scatterplot(data=base_dados, x='years_of_experience', y='salary', color='#E87C1D')
plt.xlabel('Tempo de experiência (anos)')
plt.ylabel('Salário (US$)')
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=base_dados, x='salary', y='years_of_experience')
```

```python
# Correlação
correlacao = base_dados.corr('pearson')
plt.figure(figsize=(10,5))
sns.heatmap(correlacao, annot=True, cbar=True)
```

```python
# Correlação
correlacao = base_dados.corr('spearman')
plt.figure(figsize=(10,5))
sns.heatmap(correlacao, annot=True, cbar=True)
```
